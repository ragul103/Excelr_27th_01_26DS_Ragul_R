##### 1.    Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
list_1=[1,4,2,3,5]
list_1
```

```python
# 1. append()
list_1.append(60)
print("After append(60):", list_1)
```

```python
# 2.pop()
list_1.pop()
print("After Pop():", list_1)
```

```python
# 3. extend()
list_1.extend([70, 80])
print("After extend([70, 80]):", list_1)
```

```python
# 4. insert()
list_1.insert(2, 25)   # Insert 25 at index 2
print("After insert(2, 25):", list_1)
```

```python
# 5. remove()
list_1.remove(80)   # Remove value 40
print("After remove(80):", list_1)
```

```python
# 6. index()
index_value = list_1.index(70)
print("Index of 70:", index_value)
```

```python
# 7. count()
count_value = list_1.count(70)
print("Count of 70:", count_value)
```

```python
# 8. sort()
list_1.sort()
print("After sort():", list_1)
```

```python
# 9. reverse()
list_1.reverse()
print("After reverse():", list_1)
```

```python
# 10. clear()
list_1.clear()
print("After clear():", list_1)
```

##### 6.	Given a 'marks' variable, print the grade:
- 90 and above: 'A'
- 75–89: 'B'
- 50–74: 'C'
- Below 50: 'Fail'

```python
marks=int(input("Enter you marks: "))
if marks>=90:
    print('A')
elif (marks>=75) and (marks <=89):
    print('B')
elif (marks>=50) and (marks <=74):
    print('C')
elif marks <50:
    print('Fail')
```

## Section C: Exploring Pandas

##### 1.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
# Create DataFrame manually (10 rows)
data = {
    "Name": ["Ravi", "Anu", "Kiran", "Priya", "Arun","Divya", "Vijay", "Sneha", "Ravi", "Arun"],
    "Age": [22, 24, 23, 25, 26, 22, 27, 24, 22, 26],
    "City": ["Chennai", "Bangalore", "Hyderabad", "Chennai", "Mumbai", "Delhi", "Chennai", "Bangalore", "Chennai", "Mumbai"],
    "Salary": [30000, 35000, np.nan, 40000, 45000, 30000, 50000, np.nan, 30000, 45000]
}
df=pd.DataFrame(data,columns=['Name','Age','City','Salary'])
df
```

```python
#Use .info()
df.info()
```

```python
#Use .describe()
df.describe()
```

```python
#Select 'Name' and 'City' columns
df[["Name", "City"]]
```

```python
# Drop 'City' column
df.drop('City',axis=1)
```

```python
# Fill any missing values in 'Salary' column with the mean
mean_salary = df["Salary"].mean()
df["Salary"] = df["Salary"].fillna(mean_salary)
df
```

```python
#Remove any duplicate rows
df.drop_duplicates(inplace=True)
df
```
