### 2. Encoding Techniques:

#### ●	Apply One-Hot Encoding to categorical variables with less than 5 categories

```python
# Select categorical columns
cat_cols = data.select_dtypes(include=['object']).columns

# One-Hot Encoding for columns with <5 unique values
low_card_cols = [col for col in cat_cols if data[col].nunique() < 5]
data1 = pd.get_dummies(data, columns=low_card_cols)  # creates binary columns
data1.head()
```

#### ●	Use Label Encoding for categorical variables. Data Exploration and Preprocessing:
#### ●	les with more than 5 categories

```python
from sklearn.preprocessing import LabelEncoder
# Label Encoding for columns with >=5 unique values
high_card_cols = [col for col in cat_cols if data[col].nunique() >= 5]

le = LabelEncoder()
for col in high_card_cols:
    data1[col] = le.fit_transform(data1[col])  # converts categories to numbers
data1.head()
```
