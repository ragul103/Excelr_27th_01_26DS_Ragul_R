```python
#Outlier Removal 
def remove_outliers_iqr(df):
    df_clean = df.copy()
    
    for col in df_clean.columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        
        # Better (optional improvement)
        Q1 = df.quantile(0.25)
        Q3 = df.quantile(0.75)
        IQR = Q3 - Q1

        df_clean = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
        
    return df_clean

df_clean = remove_outliers_iqr(df)
df_clean
```

### Task 2: Dimensionality Reduction with PCA:
#### 1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.


```python
from sklearn.preprocessing import StandardScaler

# Separate features (if target exists, drop it)
X = df_clean.drop("Type", axis=1)

# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### 2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# Explained variance
explained_var = pca.explained_variance_ratio_
```

```python
#Scree plot
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot")
plt.show()
```

### 3.	Transform the original dataset into the principal components

```python
# Select optimal components (example: 2)
pca = PCA(n_components=2)
X_pca_final = pca.fit_transform(X_scaled)

print(X_pca_final[:5])
```

### Task 3: Clustering with Original Data:
#### 1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.


```python
from sklearn.cluster import KMeans

# Apply KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
labels_original = kmeans.fit_predict(X_scaled)
```

#### 2.	Visualize the clustering results using appropriate plots.

```python
#scatter plot
plt.scatter(X_scaled[:,0], X_scaled[:,1], c=labels_original)
plt.title("Clustering on Original Data")
plt.show()
```

#### 3.	Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index.


```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Silhouette Score
sil_score = silhouette_score(X_scaled, labels_original)

# Davies-Bouldin Index
db_score = davies_bouldin_score(X_scaled, labels_original)

print("Silhouette Score:", sil_score)
print("Davies-Bouldin Index:", db_score)
```

### Task 4: Clustering with PCA Data:
#### 1.	Apply the same clustering algorithm to the PCA-transformed dataset.



```python
#Apply Kmeans
kmeans_pca = KMeans(n_clusters=3, random_state=42)
labels_pca = kmeans_pca.fit_predict(X_pca_final)
```

#### 2.	Visualize the clustering results obtained from PCA-transformed data.


```python
plt.scatter(X_pca_final[:,0], X_pca_final[:,1], c=labels_pca)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clustering on PCA Data")
plt.show()
```

#### 3.	Compare the clustering results from PCA-transformed data with those from the original dataset.

```python
# Silhouette Score
sil_score = silhouette_score(X_pca_final, labels_pca)

# Davies-Bouldin Index
db_score = davies_bouldin_score(X_pca_final, labels_pca)

print("Silhouette PCA:", sil_score)
print("DB PCA:", db_score)
```
